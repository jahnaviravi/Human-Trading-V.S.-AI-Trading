```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew, kurtosis
```

```python
# List of stock tickers
tickers = ['CMG', 'BKNG', 'MCD', 'MC.PA', 'NKE']

# Fetch historical data for all tickers from 2014 to 2024
stock_data = {}
for ticker in tickers:
    stock_data[ticker] = yf.download(ticker, start='2014-01-01', end='2024-01-01')

# Convert the dictionary to apandas DataFrame
stock_data = pd.concat(stock_data, axis=1)
stock_data
```

```python
# Loop through each stock and perform analysis
for ticker in tickers:
    print(f"\n=== {ticker.upper()} Analysis ===")

    # Select the current stock data
    stock_data_ticker = stock_data[ticker]

    # Plotting the adjusted close price
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data_ticker['Adj Close'], color='blue')
    plt.title(f'{ticker.upper()} Adjusted Close Price (2014-2024)')
    plt.xlabel('Date')
    plt.ylabel('Adjusted Close Price')
    plt.grid(True)
    plt.show()

    # Plotting daily percentage change
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data_ticker['Adj Close'].pct_change(), color='red')
    plt.title(f'Daily Percentage Change of {ticker.upper()} Stock (2014-2024)')
    plt.xlabel('Date')
    plt.ylabel('Percentage Change')
    plt.grid(True)
    plt.show()

    # Plotting rolling mean and standard deviation
    plt.figure(figsize=(12, 6))
    rolling_mean = stock_data_ticker['Adj Close'].rolling(window=30).mean()
    rolling_std = stock_data_ticker['Adj Close'].rolling(window=30).std()
    plt.plot(stock_data_ticker['Adj Close'], color='blue', label='Original')
    plt.plot(rolling_mean, color='red', label='Rolling Mean') 
    plt.plot(rolling_std, color='green', label='Rolling Std')
    plt.title(f'Rolling Mean & Standard Deviation of {ticker.upper()} Stock (2014-2024)')
    plt.xlabel('Date')
    plt.ylabel('Adjusted Close Price')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Statistical Analysis
    daily_returns = stock_data_ticker['Adj Close'].pct_change()
    average_daily_return = daily_returns.mean()
    std_dev_daily_return = daily_returns.std()
    annualized_average_return = average_daily_return * 252
    annualized_volatility = std_dev_daily_return * np.sqrt(252)
    
    # Calculate Skewness and Kurtosis
    skewness = skew(daily_returns.dropna())
    kurt = kurtosis(daily_returns.dropna())

    # Calculate Value at Risk (VaR) at 95% confidence level
    confidence_level = 0.05
    VaR_95 = daily_returns.quantile(confidence_level)

    print(f"Statistical Analysis ({ticker.upper()}):")
    print(f"Average Daily Return: {round(average_daily_return, 4)}")
    print(f"Daily Standard Deviation of Returns: {round(std_dev_daily_return, 4)}")
    print(f"Annualized Average Return: {round(annualized_average_return, 4)}")
    print(f"Annualized Volatility: {round(annualized_volatility, 4)}")
    print(f"Skewness: {round(skewness, 4)}")
    print(f"Kurtosis: {round(kurt, 4)}")  # Apply round to the result of kurtosis function
    print(f"Value at Risk (VaR) at {confidence_level * 100}% confidence level: {round(VaR_95, 4)}")

    # Histogram of Daily Returns
    plt.figure(figsize=(12, 6))
    plt.hist(daily_returns[1:], bins=50, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of Daily Returns ({ticker.upper()})')
    plt.xlabel('Daily Returns')
    plt.ylabel('Frequency')
    plt.show()
    
    # Bollinger Bands
    plt.figure(figsize=(12, 6))
    rolling_mean = stock_data_ticker['Adj Close'].rolling(window=20).mean()
    stock_data_ticker = stock_data_ticker.copy()  # Create a copy of the DataFrame
    stock_data_ticker.loc[:, 'Rolling Mean'] = rolling_mean
    stock_data_ticker.loc[:, 'Upper Band'] = rolling_mean + 2 * stock_data_ticker['Adj Close'].rolling(window=20).std()
    stock_data_ticker.loc[:, 'Lower Band'] = rolling_mean - 2 * stock_data_ticker['Adj Close'].rolling(window=20).std()
    plt.plot(stock_data_ticker['Adj Close'], color='blue', label='Adjusted Close Price')
    plt.plot(stock_data_ticker['Rolling Mean'], color='red', label='Rolling Mean')
    plt.plot(stock_data_ticker['Upper Band'], color='green', label='Upper Band')
    plt.plot(stock_data_ticker['Lower Band'], color='orange', label='Lower Band')
    plt.title(f'Bollinger Bands for {ticker.upper()} Stock (2014-2024)')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # MACD (Moving Average Convergence Divergence)
    plt.figure(figsize=(12, 6))
    stock_data_ticker['12-Day EMA'] = stock_data_ticker['Adj Close'].ewm(span=12, min_periods=0, adjust=False).mean()
    stock_data_ticker['26-Day EMA'] = stock_data_ticker['Adj Close'].ewm(span=26, min_periods=0, adjust=False).mean()
    stock_data_ticker['MACD'] = stock_data_ticker['12-Day EMA'] - stock_data_ticker['26-Day EMA']
    stock_data_ticker['Signal Line'] = stock_data_ticker['MACD'].ewm(span=9, min_periods=0, adjust=False).mean()
    plt.plot(stock_data_ticker.index, stock_data_ticker['MACD'], label='MACD', color='blue')
    plt.plot(stock_data_ticker.index, stock_data_ticker['Signal Line'], label='Signal Line', color='red')
    plt.title(f'MACD for {ticker.upper()} Stock (2014-2024)')
    plt.xlabel('Date')
    plt.ylabel('MACD')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Calculate short-term (50-day) and long-term (200-day) moving averages
    stock_data_ticker['SMA50'] = stock_data_ticker['Adj Close'].rolling(window=50).mean()
    stock_data_ticker['SMA200'] = stock_data_ticker['Adj Close'].rolling(window=200).mean()

    # Initialize positions and signals
    position = 0  # 1 for long position, -1 for short position, 0 for no position
    signals = []

    # Generate trading signals based on moving average crossover strategy
    for i in range(len(stock_data_ticker)):
        if stock_data_ticker['SMA50'].iloc[i] > stock_data_ticker['SMA200'].iloc[i] and position != 1:
            signals.append(1)  # Buy signal
            position = 1
        elif stock_data_ticker['SMA50'].iloc[i] < stock_data_ticker['SMA200'].iloc[i] and position != -1:
            signals.append(-1)  # Sell signal
            position = -1
        else:
            signals.append(0)  # No signal

    # Add signals to the DataFrame
    stock_data_ticker['Signals'] = signals

    # Plotting the stock prices and trading signals
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data_ticker['Adj Close'], color='blue', label=f'{ticker.upper()} Adj Close Price')
    plt.plot(stock_data_ticker['SMA50'], color='red', label='50-Day SMA')
    plt.plot(stock_data_ticker['SMA200'], color='green', label='200-Day SMA')

    # Plot buy signals
    plt.plot(stock_data_ticker[stock_data_ticker['Signals'] == 1].index,
             stock_data_ticker['SMA50'][stock_data_ticker['Signals'] == 1],
             '^', markersize=10, color='g', lw=0, label='Buy Signal')

    # Plot sell signals
    plt.plot(stock_data_ticker[stock_data_ticker['Signals'] == -1].index,
             stock_data_ticker['SMA50'][stock_data_ticker['Signals'] == -1],
             'v', markersize=10, color='r', lw=0, label='Sell Signal')

    plt.title(f'{ticker.upper()} Stock Price with Moving Averages and Trading Signals')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()
```
