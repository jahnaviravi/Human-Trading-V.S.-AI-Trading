```python
import yfinance as yf
import pandas as pd

# Fetch the constituent stocks of the S&P 500 index
sp500 = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
sp500_symbols = sp500['Symbol'].tolist()

# Define your criteria for selecting top stocks
# You can adjust the weights according to your preference
weights = {
    'Volume': 0.4,
    'Volatility': 0.3,
    'Price_Movement': 0.3
}

top_5_stocks = []

# Loop through each stock symbol in the S&P 500
for symbol in sp500_symbols:
    try:
        # Fetch historical data for the stock
        stock_data = yf.download(symbol, start="2023-01-01", end="2023-12-31")
        
        # Calculate volatility and price movement
        stock_data['Volatility'] = stock_data['Close'].pct_change().rolling(window=20).std() * (252 ** 0.5)
        stock_data['Price_Movement'] = stock_data['Close'].pct_change() * 100
        
        # Calculate average volume
        average_volume = stock_data['Volume'].mean()
        
        # Calculate average volatility
        average_volatility = stock_data['Volatility'].mean()
        
        # Calculate average price movement
        average_price_movement = stock_data['Price_Movement'].mean()
        
        # Calculate score based on weights
        score = (average_volume * weights['Volume']) + (average_volatility * weights['Volatility']) + (average_price_movement * weights['Price_Movement'])
        
        # Append stock symbol and score to the list
        top_5_stocks.append((symbol, score))
    except:
        pass

# Sort the stocks by their scores in descending order and select the top 5
top_5_stocks.sort(key=lambda x: x[1], reverse=True)
top_5_stocks = top_5_stocks[:5]

print("Top 5 Stocks to Trade from S&P 500 INDEX:")
for stock in top_5_stocks:
    print(stock[0], "Score:", stock[1])
```
